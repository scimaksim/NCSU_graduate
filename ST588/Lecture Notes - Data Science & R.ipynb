## Lecture 2 - Common Data Objects <a class="anchor" id="common_data_objects"></a>

There are 5 major data structures:
1. Atomic vector (1-D)
2. Matrix (2-D)
3. Array
4. Data frame (2-D)
5. List (1-D)



#### Atomic vector (one-dimensional) <a class="anchor" id="atomic_vector"></a>
Most basic. A vector is a set of elements with some kind of ordering (first element, second element, third element, etc.) where every element is of the same type (all numbers, all character strings, all logicals, etc.). By themselves, vectors are not generally useful because we often have more than one type of interesting variable. However, they are often the building blocks of matrices or data frames, where each column represents a vector.

We use the function `c()` to **c**ombine values. 

````r
# Combine values into 1-dimensional atomic vector
# All elements are of the same type
x <- c(1, 3, 10, -20, sqrt(2))
y <- c("cat", "dog", "bird", "floor")

x
y
````

We can use a short-hand to make sequences. The output is a vector.

````r
# Short-hand for sequences and element-wise division
# This divides each integer (1-20) by 20
1:20 / 20
````

More involved sequences can be created using the `seq()` ("sequence") function. Again, the output is a vector.

````r
# Customize the sequence
seq(from = 1, to = 10, by = 2)
seq(to = 5)
# help(seq)
````

There are also many ways to do random number generation. One such way is to use `runif` (pronounced "r-unif").

````r
# Generate 4 random numbers between 0 and 1
runif(4, min = 0, max = 1)
# help(runif)
````

#### Matrix (two-dimenstional) <a class="anchor" id="matrix"></a>

A matrix is a **collection of vectors** of the **same type and length**. They can be useful for data, but not entirely, since everything in the matrix must be of the same type. Usually, our data will have some variables which are descriptive and some variables which are numeric.

````r
# Create two vectors
x <- rep(0.2, times = 6)
y <- c(1, 3, 4, -1, 5, 6)
# help(rep) 
# `rep` replicates (repeats) the value in x: rep(x, ...)

x
y

# Confirm that x and y are numeric
is.numeric(x)
is.numeric(y)

# Check length of x and y
length(x)
length(y)

# Combine x and y into a matrix
# help(matrix)
matrix(c(x, y), ncol = 2)
matrix(c(x, y), nrow = 2, byrow = TRUE)
````

````r
# Create character vectors
x <- c("Hi", "There", "!"); y <- c("a", "b", "c"); z <- c("One", "Two", "Three")
x
y
z
is.character(x)

# Create matrix
matrix(c(x, y, z), nrow = 3)
matrix(c(x, y, z), nrow = 3, byrow = TRUE)
````

#### Data Frame <a class="anchor" id="data_frame"></a>

A data frame is a **collection of vectors** of the **same length** (not type). Data frames are a great object for data. Usually, the numbers of rows represents the numbers of observations and the numbers of columns represents the number of variables. 

````r
# Create three vectors of different types
x <- c("a", "b", "c", "d", "e", "f")
y <- c(1, 3, 4, -1, 5, 6)
z <- 10:15

# Create a data frame
data.frame(x, y, z)

# Change the names of our columns
data.frame(char = x, data1 = y, data2 = z)
````

##### List <a class="anchor" id="list"></a>
Lists are very flexible. They are considered one-dimensional because have an **ordering**, but they can contain **different element types** (e.g. a matrix in the first element, a vector in the second, a data frame in the third, another list in the fourth, etc.). 

````r
# Create a list
list("Hi", 1:3, rnorm(2), c("!", "?"))
````

### Basic data manipulation <a class="anchor" id="data_manipulation"></a>

For a given data set, we may want to grab a single value, a single column, multiple columns, a single row, multiple rows, or the values of an attribute. 

* Atomic vectors: x[]
* Matrices: x[ , ]
* Data frames: x[ , ] or `x$name`
* Lists: x[[]] or `x$name`

#### Manipulate atomic vectors <a class="anchor" id="manipulate_vectors"></a>

We access the elements of a vector using square brackets, [].

````r
# Print the built-in vector "letters"
letters

# Access the tenth element of "letters"
letters[10]

# Access the first four elements of "letters"
letters[1:4]

# Access a set of arbitrary letters
letters[c(13, 1, 11, 19, 9, 13)]

# Feed a vector to the square bracket
x <- c(14, 9, 11, 9, 6, 15, 18, 15, 22); letters[x]
````

#### Manipulate matrices <a class="anchor" id="manipulate_matrices"></a>

Matrix elements are accessed using square brackets with a comma in-between - **[row, column]**.

````r
# Print a basic matrix
mat <- matrix(c(1:4, 20:17), ncol = 2)
mat

# Access specific matrix element
mat[2, 2]

# Access first column
mat[ , 1]

# Access second row
mat[2, ]

# Access rows 2-4 and in the first column
mat[2:4, 1]

# Access rows 2 and 4
mat[c(2, 4), ]
````

We can also use give columns names and and use these names for access. View `help(matrix)` for details. There is an option, `dimnames`, which accepts a list. The first element of the list is row names and the second element of the list is column names.

````r
# Assign column names to matrix
mat <- matrix(c(1:4, 20:17), ncol = 2, 
              dimnames = list(NULL, 
                              c("First Column", "Second Column")))

mat

mat[ , "First Column"]
````

We can also look at the `str()` and `attributes()` of a matrix. Recall that all elements in a matrix must of the same type. 

````r
# View the structure of mat
str(mat)

# View the metadata of mat. 
# Recall that attributes() returns a named list
attributes(mat)
````

#### Manipulate data frames <a class="anchor" id="manipulate_data_frames"></a>

Data frames are two-dimensional, so we can use access conventions which are similar to those of matrices. 

````r
# View the built-in iris data set
iris

# View the structure of the iris data set, which is
# of type data.frame
str(iris)

# List first four rows and columns 2-4
iris[1:4, 2:4]

# List the complete first row of iris
iris[1, ]

# List specific columns using variable names
iris[ , c("Sepal.Length", "Species")]

# Access column contents (returns a vector)
iris$Sepal.Length

# Access specific element in a column
iris$Sepal.Length[52]
````

#### Manipulate lists <a class="anchor" id="manipulate_lists"></a>

Remember that a list is one-dimensional and each element can be different (i.e. a matrix for the first element, data frame for the second, another list for the third, etc.). **Use double brackets** to access the elements of a list.

````r
# Create a list
mylist <- list("HI", c(10:20), 1)
mylist

# Access the second element in mylist
mylist[[2]]

# Access the fourth and fifth values 
# in the second element of mylist
mylist[[2]][4:5]
````

If your list elements have names to them (i.e. named list), we can access them with the `$` sign.

````r
# Create a named list
my_named_list <- list("HI", c(10:20), 1)
str(my_named_list) # note that nothing follows the $ sign (unnamed elements)

# Assign names
my_named_list <- list(First = "HI", Second = c(10:20), Third = 1)

my_named_list
str(my_named_list) # note that there are now names following the $ sign

my_named_list$Second
````

Data frames are just lists where each element is an equal-length vector. Lists have `$` signs and data frames do as well.

````r
# List the structure of iris
str(iris)

# Optionally, access the second list in iris using double-square brackets
# However, it's easier to just use the name
iris[[2]]
iris$Sepal.Width
````

#### Accessing and manipulating attributes <a class="anchor" id="accessing_attributes"></a>

Frequently, we may want to change the names of our columns.

````r
# Return the name of the columns in the iris data set
attributes(iris)$names # a vector of length 5

# Overwrite a given element
attributes(iris)$names[1] <- "Sepal_Length"
attributes(iris)$names
````

### Coercion <a class="anchor" id="coercion"></a>

Generally, R tries to coerce data **from less flexible** to more flexible. 

In terms of flexibility, characters > double > integer > logical. That is, **characters are the most flexible** and logicals are the least flexible. Generally, we can convert a logical to an integer, but we cannot convert a character to a double. 

- logical (least flexible)
- integer
- double
- character (most flexible)

Implicit coercion means that, in the background, your programming language is changing things (e.g. their type).

````r
# Example of implicit coercion.
# Implicitly coerce number to string
# "c" is a way to create vectors and stands for "combine".
# Since vectors have to have all elements of the same type,
# R tries to convert 10 to a character.
c("hi", 10)

# R doesn't know how to convert "hi" to a logical,
# but it knows how to convert "TRUE" to a character
c(TRUE, "hi")
````

Frequently, we perform explicit coercion with `as.` functions

````r
# Coerce TRUE/FALSE to numeric.
# TRUE/FALSE are a special type in R, and any time we
# do math on a TRUE/FALSE, R converts a "TRUE" to 1 and
# a "FALSE" to 0.
c(TRUE, FALSE) + 0

# Explicitly coerce with as.

as.numeric(c(TRUE, FALSE, TRUE))

# Implicit coercion, since we aren't converting the
# TRUE/FALSE values to 0's and 1's ourselves. We could
# do it explicitly with as.numeric()
mean(c(TRUE, FALSE, TRUE))

# Implicitly coerce "TRUE" to 1 using the c function 
# because we're combining it with other numerical (double) values
# and then explicitly coerce to the string "1". 
as.character(c(1, 2, 3.5, TRUE))
````

### Compound statements <a class="anchor" id="compound_statements"></a>

We can perform a logical `AND` and a logical `OR` using `&` and `|`, respectively. 

Doubles (`&&` and `||`) are used to compare just the first value. 

````r
# Set a random number generator seed to 
# get the same set of numbers every time we run this code
set.seed(3)

# Generate 10 random uniform values between 0 and 1
x <- runif(n = 10, min = 0, max = 1)
x

# Do a compound logical against each value and return
# TRUE if a value is less than 0.25 and 
# another TRUE if a value is greater than 0.75

(x < 0.25) | (x > 0.75)

# Compare just the first value

(x < 0.25) || (x > 0.75)
````

**A:** When you do math on a `TRUE`, it is conveted to a 1.

````r
c(TRUE, 1, "hi")

c(c(TRUE, 1), "hi")
````

#### tbl_df() <a class="dplyr_basic_commands" id="tbl_df"></a>

Tibbles are special types of data frames with better printing. If we have a data frame, we can convert it to a tibble just by "wrapping" it using the **tbl_df()** function. This essentially converts data frames to ones with better printing. Data frames and tibbles can both contain data sets with variables (columns) of differing types.

`as_tibble()` is now the preferred function to use rather than tbl_df(). These functions just convert the data frame to a tibble. Tibbles have better printing properties and don't simplify what gets returned when returning only one column like a data frame does (ex: `iris[, 1]` returns a vector but `as_tibble(iris)[,1]` returns a tibble with one column).

````r
# Data frame returns a vector
iris[, 1]

# Tibble with one column
as_tibble(iris)[,1]
````

#### mutate() <a class="dplyr_basic_commands" id="mutate"></a>

This function adds new columns to the current data frame, behaving much like `cbind()`. It features the syntax 

```
mutate(data, newVarName = functionOfData, newVarName2 = functionOfData, ...)
```

As an example, consider a data set that contains movie ratings.

````r
# Create an average rottentomatoes score variable and an
# average metacritic score variable
# (does not overwrite original fandango)
# and make them easily viewable

fandango %>% 
  mutate(avgRotten = (rottentomatoes + rottentomatoes_user)/2,
        avgMeta = (metacritic_norm + metacritic_user_nom)/2) %>%
  select(film, year, avgRotten, avgMeta, everything())
````

With `mutate()`, we can also use statistical functions, such as `mean()` and `sd()`, and we can create summaries for group using `group_by()`. The [dplyr cheatsheet](https://www.rstudio.com/wp-content/uploads/2015/02/data-wrangling-cheatsheet.pdf) lists other commonly used functions.

````r
# Take a value and keep adding to it using "cumsum"

fandango %>%
  select(rottentomatoes) %>%
  mutate(cumulativeSum = cumsum(rottentomatoes))
````

#### Conditional execution  <a class="anchor" id="conditional_execution"></a>

We may want to execute statements conditionally to create a variable.

```
if (condition) {
  then execute code
}

# if then else
if (condition) {
  execute this code
} else {
  execute this code
}
```

or

```
# Or more if statements
if (condition) {
  then execute code
} else if (condition2) {
  execute this code
} else if (condition3) {
  execute this code
} else {
  # if no conditions met
  execute this code
}
```

````r
# Example using a built-in data set, "airquality", which measures
# daily air quality in New York from May (day 1) to September (day 153) in 1973.
# It is a data frame (not a tibble). Tibbles print nicely
# and do not simplify when you subset a single column.

# Create a tibble from the data frame
myAirquality <- as_tibble(airquality)
myAirquality
````

We can create a wind category variable with the following conditions:

* high wind days (wind $\geq$ 15 mph)
* windy days (10 mpg $\leq$ wind < 15 mph)
* lightwind days (6 mph $\leq$ wind < 10 mph)
* calm days (wind $\leq$ 6 mph)

Our issue is that `if(condition)` only takes in a single comparison and looping in R is slow. Instead, we can use the vectorized function `if_else()` from the dplyr package to work on an entire vector. 

The syntax for `if_else()` is 

```
if_else(condition, true, false)
```

* **condition** is a vector of TRUE/FALSE
* **true** is what to do when TRUE occurs
* **false** is what to do when FALSE occurs

````r
# Create Status variable with wind conditions

myAirquality <- myAirquality %>%
  mutate(Status = if_else(Wind >= 15, "HighWind", 
                         if_else(Wind >= 10, "Windy",
                                if_else(Wind >= 6, "LightWind", "Calm"))))

myAirquality
````

## Lecture 5 - Reshaping data using tidyr  <a class="anchor" id="reshaping_data"></a>

We often want to switch data from a wide format to a long format. Two functions in **tidyr** allow us to do this - `pivot_longer()` and `pivot_wider()`.

### pivot_longer()  <a class="anchor" id="pivot_longer"></a>

This function lengthens data by increasing the number of rows and decreasing the number of columns. This is often the preferred format for analysis. 

````r
# Read temperature data on a few different cities in "wide" format.
# Each row is 7 observations on each city (wide format),
# rather than one row that corresponds to one observation.

tempsData <- read_table2(file = "https://www4.stat.ncsu.edu/~online/datasets/cityTemps.txt")
tempsData
````

To turn this into a long format, we would need 3 variables:

1. City
2. Temperature
3. Day

If we use `pivot_longer()` to switch to long form, we need the call arguments

- **cols** - which columns we want to take from wide format to long format (cols = 2:8)
- **names_to** - news name(s) for columns (names_to = "day")
- **values_to** - new name(s) for data values (values_to = "temp")

````r
# Switch data from wide to long format

tempsData %>% 
  pivot_longer(cols = 2:8, names_to = "day", values_to = "temp")
````

### pivot_wider()  <a class="anchor" id="pivot_wider"></a>

This function widens data by increasing the number of columns and decreasing the number of rows. 

We need two arguments:

- **names_from** - column(s) to get the names used in the output columns (names_from = "day")
- **values_from** - column(s) to get the cell values from (values_from = "temp")

````r
#

newTempsData <- tempsData %>%
  pivot_longer(cols = sun:sat, names_to = "day", values_to = "temp")

newTempsData %>%
  pivot_wider(names_from = "day", values_from = "temp")
````

### separate()  <a class="anchor" id="separate"></a>

We may have a column such as "date" with entries such as "03/21/2019". We may want to separate values within this column into multiple columns. We can do this using the `separate()` function.  

````r
# Obtain Chicago data set

chicagoData <- read_csv("https://www4.stat.ncsu.edu/~online/datasets/Chicago.csv")
chicagoData
````

````r
# Separate "date" column into month, day, and year.
# convert = TRUE converts char to int.
# Default behavior is to remove the original variable,
# so remove = FALSE inhibits that.

chicagoData %>% 
  separate(date, c("Month", "Day", "Year"), 
           sep = "/", convert = TRUE, remove = FALSE)
````

### unite()  <a class="anchor" id="unite"></a>

Alternatively, we may want to unite values in multiple columns into one column. We can do this with the `unite()` function. 

````r
# Combine columns. Month and Day no longer exist in the data set 
# because remove = TRUE is the default, and we did not overwrite it
# within the unite() function. 

chicagoData %>% 
  separate(date, c("Month", "Day", "Year"), 
           sep = "/", convert = TRUE, remove = FALSE) %>%
  unite(MonthDay, Month, Day, sep = "-")
````

### Writing our own functions  <a class="anchor" id="writing_functions"></a>

Suppose we wish to create a function called `standardize()`. This function must 

1. Take a vector of values
2. Subtract the mean
3. Divide by standard deviation

For a given value _i_,

$$
\frac{value[i] - mean(value)}{sd(value)}
$$

````r
# Code for our custom function, standardize()

standardize <- function(vector) {
    return((vector - mean(vector)) / sd(vector))
}
````

````r
# Use our custom function

data <- runif(5)
data

result <- standardize(data)
result

# Check that the mean is 0 and the SD is 1

mean(result)
sd(result)
````

Let's now customize our function.

````r
# Code for our custom function, standardize().
# We may want to make centering or scaling optional.
# Implicitly, we are saying "center" and "scale" 
# need to be a TRUE or FALSE.

standardize <- function(vector, center, scale) {
    if (center) {
        vector <- vector - mean(vector)
    }
    if (scale) {
        vector <- vector / sd(vector)
    }
    return(vector)
}
````

````r
# Use our new function by providing argument for center and scale
 
result <- standardize(data, center = TRUE, scale = TRUE)
result

result <- standardize(data, center = FALSE, scale = TRUE)
result
````

We can give users options to only center or to only scale by setting default values.

````r
# Code for our custom function, standardize(), 
# with default values.

standardize <- function(vector, center = TRUE, scale = TRUE) {
    if (center) {
        vector <- vector - mean(vector)
    }
    if (scale) {
        vector <- vector / sd(vector)
    }
    return(vector)
}
````

````r
# Now, the two calls below are equivalent.
 
result <- standardize(data, center = TRUE, scale = TRUE)
result

result <- standardize(data)
result
````

Eventually, we may want to return more than a single vector. We may also want to return the mean and the standard deviation of the original data.

To return more than one thing, we will need to return a list. Lists are one-dimensional, and each element can be of a different type. 

````r
# Code for our custom function, standardize(), 
# with additional returns

standardize <- function(vector, center = TRUE, scale = TRUE) {
    mean <- mean(vector)
    stdev <- sd(vector)
    if (center) {
        vector <- vector - mean
    }
    if (scale) {
        vector <- vector / stdev
    }
    return(list(vector, mean, stdev))
}
````

````r
# Now, result returns three things
 
result <- standardize(data)
result

# To get only the second result

result[[2]]
````

Running `result[[2]]` is not pleasing to the eye, so we can enhance our custom function.

````r
# Code for our custom function, standardize(), 
# with named returns.

standardize <- function(vector, center = TRUE, scale = TRUE) {
    mean <- mean(vector)
    stdev <- sd(vector)
    if (center) {
        vector <- vector - mean
    }
    if (scale) {
        vector <- vector / stdev
    }
    return(list(result = vector, mean = mean, sd = stdev))
}
````

````r
# Now, our returns are far more informative
 
result <- standardize(data, center = TRUE, scale = TRUE)
result

result$sd
````

### Infix functions <a class="anchor" id="infix_functions"></a>

An infix function is a function that goes between arguments (as opposed to prefix). Common examples include `::`, `$`, `^`, `*`,  etc. Others use `%symbol%`, such as `%*%` for matrix multiplication. 

````r
# Examples

mean(3:5) # prefix

3 + 5 # infix

`+`(3, 5) # used as a prefix function

cars <- as.matrix(cars)
t(cars) %*% cars # transpose

`%*%`(t(cars), cars)
````

We can write our own infix functions by naming it with backticks.

````r
# Create our own infix function that pastes together
# two strings with no separator

`%+%` <- function(a, b) paste0(a, b)
"new" %+% " string"
````

Normally, rules of precedence work as follows: 

````r
x <- y <- 2 # works as expected

x <- y = 2 # error, since <- has higher precedence

x = y <- 2 # this will work!
````

However, user-defined infix operators are evaluated left to right!

````r
`%-%` <- function(a, b) {
    paste0("(", a, " %-% ", b, ")")
}

"a" %-% "b" %-% "c"

# The code above is interpreted as 
# `%-%`(`%-%`("a", "b"), "c")
````
